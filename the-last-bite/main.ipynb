```python
from langchain.agents import create_agent
from langchain.tools import tool
from langgraph.checkpoint.memory import InMemorySaver
!pip install tavily-python
from tavily import TavilyClient
from dotenv import load_dotenv
load_dotenv()
!pip3 install --upgrade tavily
```

```python
# To install: pip install tavily-python
from tavily import TavilyClient
client = TavilyClient()


@tool(description="You need to use this tool to search the web for information. Use this tool when you need to find information that is not available in the current conversation or when you want to verify facts. Always use this tool when you need to search the web.")
def search(query: str) -> str:
    response = client.search(
        query=query,
        search_depth="advanced"
    )
    print("The web search response is: ", response)
    print("\n\n\n\n\n\n\n\n")
    return response
```

```python

a = input("Enter the count of ingredients:  ")
b = [input(f"Enter ingredient {i+1}: ") for i in range(int(a))]

config = { "configurable": { "thread_id": "1" } }
agent = create_agent(
    tools=[search],
    checkpointer=InMemorySaver(),
    model="google_genai:gemini-3-flash-preview",
   
    system_prompt=
    """

          You are an expert chef, who has the expertise to create recipes based on the ingredients provided by the user.
          You have the access to the tool web_search, where you can search for the recipes based on the ingredients provided by the user. 
          The you need to store the recipes which has been searched in the memory, so that you can refer to it later if needed.
          The user will provide you with the list of ingredients, and you need to search for the recipes based on those ingredients, and then provide the recipes to the user.
          Do not hallucinate the recipes, make sure its based on the search results, and if you are not able to find any recipe based on the
      search results, then you can say that you are not able to find any recipe based on the ingredients provided by the user.


        OUTPUT FORMAT INSTRUCTION:
        When you are providing the recipes to the user, make sure to provide the recipes in the following format:
        Recipe Name: <Name of the recipe>
        Ingredients: <List of ingredients required for the recipe>
        Instructions: <Step by step instructions to make the recipe>    
    """
)
response = agent.invoke(
    {
        "messages": [
            {"role": "user", "content": f"Find a recipe based on these ingredients: {', '.join(b)}"}
        ]
    },
    config=config
)
print(response)
```
